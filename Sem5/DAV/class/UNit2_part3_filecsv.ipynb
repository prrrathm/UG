## use the chunksize argument to pandas.read_csv gives  an iterator over DataFrames, rather than one single DataFrame. Each DataFrame is the next chunksize lines of the CSV:

```python
irischunk=pd.read_csv('Iris.csv',header=0,chunksize=5)
print(irischunk)
```

```python
for piece in irischunk:
  print(piece)
```

## processing each chunk and writing to CSV

```python
irischunk=pd.read_csv('Iris.csv',header=0,chunksize=10)
tot = pd.Series([]); count=0
for piece in irischunk:
  #print(piece)
  tot = tot.add(piece['Species'].value_counts(), fill_value=0)
  if count==16:
    break
  count+=1
tot = tot.sort_values(ascending=False)
print(tot)

```

```python
tot
```

```python
tot.sort_index(ascending=False)
```

```python
tot.to_csv('totalIris.csv')  #writig series/dataframe to the file
```

# Web APIs for data using 'requests' which:


*   used for making HTTP requests in Python. 
*   provides simple mechanism by abstracting the complexities of making HTTP requests for interacting with services and consuming data in the application.



## provides two powerful methods: get() and POST()
The GET method is used to get or retrieve data from a specified resource.  by using requests.get() which returns a status_code to indicate the status of request. Code :

*   200:  means request was successful and the server responded with the data you were requesting.
*   404: No source found

*   204: successful but no data and many more


## Let us try to retrieve 30 GitHub issues for PANDAS on GitHub

```python
import requests
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
response = requests.get(url)
response
```

## status_code returned is 200, which means request was successful and the server responded with the data that was requested.

```python
type(response)
```

## used Headers to get useful information like content type of the response payload and a time limit on how long to cache the response.

```python
response.headers
```

## converting the collected responses into data (excluding comments) dictionaty using json() method. Each element in data is a dictionary containing all of the data found on a GitHub issue page (except for the comments)

```python
data = response.json()
data[1]['title']
```

```python
type(data)
```

```python
issues = pd.DataFrame(data, columns=['number', 'title',
                                     'labels', 'state'])
issues
```
