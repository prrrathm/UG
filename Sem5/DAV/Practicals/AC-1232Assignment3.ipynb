Create a dataframe having at least 3 columns and 50 rows to store numeric data generated using a randomfunction. Replace 10% of the values by null values whose index positions are generated using randomfunction.

Do the following:

1. Identify and count missing values in a dataframe.
2. Drop the column having more than 5 null values.
3. Identify the row label having maximum of the sum of all values in a row and drop that row.
4. Sort the dataframe on the basis of the first column.
5. Remove all duplicates from the first column.
6. Find the correlation between first and second column and covariance between second and thirdcolumn.
7. Detect the outliers and remove the rows having outliers.
8. Discretize second column and create 5 bins

```python

Q1DF = pd.DataFrame(
    data=np.random.randn(150).reshape(50, 3),
    columns=['col1', 'col2', 'col3']
)
for i, j in zip(np.random.randint(0, 50,(15)), np.random.randint(0, 3, (15))):
    Q1DF.iloc[i, j] = np.nan
    # print(i,j)
Q1DF
```

```python
# Identify missing values in a dataframe.
Q1DF.isna().values
```

```python
# Count missing values in a dataframe.
# Q1DF.loc['col1']
Q1DF.isna().sum()
```

```python
# Drop the column having more than 5 null values.
[Q1DF.drop(x, axis=1, inplace=True)
 for x, y in zip(
    Q1DF.isna().sum().index,
    Q1DF.isna().sum().values
)
    if y > 5]

Q1DF
```

```python
# Identify the row label having maximum of the sum of all values in a row and drop that row.

[Q1DF.drop(x,axis=1) for x in Q1DF.sum(1) if x==Q1DF.sum(1).max]
Q1DF
```

```python
# 4. Sort the dataframe on the basis of the first column.

Q1DF.sort_values(by=['col1'])
```

```python
# 5. Remove all duplicates from the first column.
Q1DF = Q1DF.iloc[: , 1:]
Q1DF
```

```python
# 6. Find the correlation between first and second column and covariance between second and third column.
Q1DF.iloc[:,0].corr(Q1DF.iloc[:,1])
```

```python
Q1DF.iloc[:,0].cov(Q1DF.iloc[:,1])
```

```python
# 7. Detect the outliers and remove the rows having outliers.
Q1DF.describe()
```

```python
Q1DF[Q1DF.iloc[:,1]> (Q1DF.iloc[:,1].mean()+2*Q1DF.iloc[:,1].std())]
```

```python
# 8. Discretize second column and create 5 bins
pd.cut(Q1DF.iloc[:,1],bins=5)
```

2: Create a DataFrame to store marks of M students for n subjects and do the following :

1. Find Average marks for each subject and add as column
2. Display average marks of each subject and ass as a new row
3. Compute descriptive statistics subject-wise.
4. Compute grade obtained by each student as per the examination policy of your course (use lambda function)
5. Find frequeency of each grade for your class
6. Find frequency of each grade obtained by each student and create a new DF as the following and set Rollno as row index of DF

```python
Q2DF = pd.DataFrame(
    np.random.randint(35, 100, (6, 5)),
    columns=['English', 'Maths', 'PE', 'Hindi', 'Science']
)
Q2DF
```

```python
# 1. Find Average marks for each subject and add as column
Q2DF['Average'] = Q2DF.mean(axis=1,numeric_only=True)
Q2DF
```

```python
# 2. Display average marks of each subject and ass as a new row
Q2DF.loc['Average'] = Q2DF.mean(numeric_only=True)
Q2DF
```

```python
# 3. Compute descriptive statistics subject-wise.
Q2DF.describe()
```

```python
# 4. Compute grade obtained by each student as per the examination policy of your course (use lambda function)
Q2DF['Grade'] = Q2DF['Average'].apply(lambda x: 'A' if x >= 80 else ('B' if (x>=60 and x<=70)else 'C'))
Q2DF
```

```python
# 5. Find frequency of each grade for your class
Q2DF['Grade'].value_counts()
```

```python
# 6. Find frequency of each grade obtained by each student and create a new DF as the following and set Rollno as row index of DF
Q2DF2 = Q2DF.copy()
Q2DF2['English'] = Q2DF2['English'].apply(
    lambda x: 'A' if x >= 80 else ('B' if (x >= 60 and x <= 70)else 'C'))

Q2DF2['Science'] = Q2DF2['Science'].apply(
    lambda x: 'A' if x >= 80 else ('B' if (x >= 60 and x <= 70)else 'C'))

Q2DF2['Maths'] = Q2DF2['Maths'].apply(
    lambda x: 'A' if x >= 80 else ('B' if (x >= 60 and x <= 70)else 'C'))

Q2DF2['PE'] = Q2DF2['PE'].apply(
    lambda x: 'A' if x >= 80 else ('B' if (x >= 60 and x <= 70)else 'C'))

Q2DF2['Hindi'] = Q2DF2['Hindi'].apply(
    lambda x: 'A' if x >= 80 else ('B' if (x >= 60 and x <= 70)else 'C'))

Q2DF2 = Q2DF2[:-1]
Q2DF2.drop(columns=["Average","Grade"],inplace=True)
Q2DF2.insert(0, 'RollNo', [1, 2, 3, 4, 5, 6])
Q2DF2
```

```python
(Q2DF2.set_index('RollNo').apply(lambda x:x.value_counts(),axis=1).fillna(0).astype(int))
```
