Do the following qs done in assignment 3 using group by and its functionalities wherever needed:

```python
# 1
exam_data = {
    "Roll Number": np.random.randint(low=0, high=50, size=10),
    "Name": [
        "Rajat",
        "Khushi",
        "Ronny",
        "Anupam",
        "Vishal",
        "Ram",
        "Utsav",
        "Anil",
        "Prakash",
        "Preeti",
    ],
    "Physics": np.random.randint(low=0, high=100, size=10),
    "Chemistry": np.random.randint(low=0, high=100, size=10),
    "Maths": np.random.randint(low=0, high=100, size=10),
}

que3df = pd.DataFrame(exam_data)
que3df
```

```python
# 1
que3df["Average"] = que3df.mean(numeric_only=True, axis=1)
que3df
```

```python
# 3
que3df.describe()
```

```python
# 4
que3df["Grade"] = que3df.apply(
    lambda x: "A1"
    if x.Average >= 90
    else "A2"
    if (x.Average >= 80 and x.Average < 90)
    else "B1"
    if (x.Average >= 70 and x.Average < 80)
    else "B2"
    if (x.Average >= 60 and x.Average < 70)
    else "C1"
    if (x.Average >= 50 and x.Average < 60)
    else "C2"
    if (x.Average >= 40 and x.Average < 50)
    else "D1"
    if (x.Average >= 30 and x.Average < 40)
    else "F",
    axis=1,
)
que3df
```

```python
# 5
que3df[["Grade", "Average"]].groupby("Grade").count()
```
